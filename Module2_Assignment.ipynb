```python
def get_employee_details(name):
    try:
        # Filter data by employee name
        employee = df[df['EmployeeName'].str.lower() == name.lower()]

        if employee.empty:
            raise ValueError(f"Employee '{name}' not found.")

        return employee
    except Exception as e:
        return str(e)
```

```python
get_employee_details('NATHANIEL FORD')
```

```python
def employee_dict(name):
    employee = get_employee_details(name)
    if isinstance(employee, str):
        return employee # Return error message
    else:
        # Convert row to dictionary
        return employee.to_dict(orient='records')[0]

# Example
employee_dict('NATHANIEL FORD')
```

```python
import csv
import os
import zipfile

def export_employee(name):
  details = employee_dict(name)
  if isinstance(details, str):
    print(details)
    return

  # Create folder if it doesn't exist
  folder_name = "Employee_Profile"
  os.makedirs(folder_name, exist_ok=True)

  # CSV file path
  csv_file = os.path.join(folder_name, "Employee_Details.csv")

  # Write CSV
  with open(csv_file, mode='w', newline='') as file:
    writer = csv.DictWriter(file, fieldnames=details.keys())
    writer.writeheader()
    writer.writerow(details)

  # Zip the folder
  zip_file = f"{folder_name}.zip"
  with zipfile.ZipFile(zip_file, 'w') as zipf:
    for root, dirs, files in os.walk(folder_name):
      for file_name in files:
        file_path = os.path.join(root, file_name)
        zipf.write(file_path, arcname=os.path.relpath(file_path, folder_name))

  print(f"Employee Details exported and zipped as {zip_file}")
```

```python
export_employee('NATHANIEL FORD')
```
